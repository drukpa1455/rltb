```python
from ray.rllib.algorithms.ppo import PPOConfig
```

```python
# 1. Configure
config = (
    PPOConfig()
    .environment("Taxi-v3")
    .rollouts(num_rollout_workers=2)
    .framework("torch")
    .training(model={"fcnet_hiddens": [ 64, 64]})
    .evaluation(evaluation_num_workers=1)
)
```

```python
# 2. Build
algo = config.build()
```

```python
# Train
for _ in range(5):
    print(algo.train())
```

```python
# Evaluate
algo.evaluate()
```
